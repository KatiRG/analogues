```python
import numpy as np
import pandas as pd
import collections
import os
```

```python
num_analogues = 20 #number of analogues searched for
```

```python
# Create dataframe from analogues file

dfS = pd.DataFrame()
dfS = pd.read_csv('analogues.tsv', delimiter=' ', index_col=0)
#dfS.set_index('Date', inplace=True)

dfS.head(10)

#dfS.index.name
```

```python
#SAVE DATE INDEX
#DF['id2'] = DF.index
dfS['dateRef'] = dfS.index
dfS['dateRef'].head(10)
```

```python
#CREATE ANALOGUE DATE TABLE ONLY
#s1.iloc[:3] #selects rows only
# df1.iloc[1:5,2:4] #selects rows, cols
dfA_raw = dfS.iloc[:, 0:20]
dfA_raw.head(10)
```

```python
dfA_raw.shape
```

```python
20*24831
```

```python
#STACK ANALOGUE DATES ACROSS COLUMNS INTO ONE COLUMN

#first remove index name of dfA_raw
dfA_raw.index.name = ""

dfA_stack = dfA_raw.stack()
dfA_stack.index.name
```

```python
dfA_stack
```

```python
#list(dfA_stack)
```

```python
#CREATE EMPTY DATAFRAME WITH NUM ROWS = NUM COLS OF dfA_raw
index = dfA_raw.shape[0] #num cols of dfA
columns = 2
dfA = pd.DataFrame(np.nan, index=range(0, dfA_raw.shape[0]), columns=['dateAnlg'])

#dfA = pd.DataFrame(np.nan, index=range(0,4), columns=['Adate'])

dfA.shape
```

```python
#REPLACE INDEX COL OF dfA WITH ORIG date_id
dfA = dfA.set_index(dfS['dateRef'])
dfA.head(10)
#dfA.shape
```

```python
#REPLICATE INDEX COL 20 TIMES PER ROW

#http://stackoverflow.com/questions/26777832/replicating-rows-in-a-pandas-data-frame-by-a-column-value
#df2 = df.loc[np.repeat(df.index.values,df.n)]
dfA = dfA.loc[np.repeat(dfA.index.values,num_analogues)]

```

```python
dfA.head(30)
```

```python
#REPLACE COL 0 WITH STACKED DATES
dfA.iloc[:,0] = list(dfA_stack)
```

```python
dfA.head(30)
```

```python
#SAVE TO TSV FILE

dfA.to_csv('analogues_reformat.tsv', sep='\t')
```
